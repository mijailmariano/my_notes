### Learning MySQL

*MySQl is a relational database management systerm or "RDBMS"*

- MySQL is one of the more popular and most commonly used RDBMS

- this means that it stores data in tables and creates a relationship between the data in the different tables

- similar to having multiple google sheets or microsoft excel spreadsheets. you have data on one sheet and use data from another. 

- however, unlike spreadsheets - relational databases like MySQL are able to manipulate millions of rows and columns rapidly and effectively

- relational databases are the most common way to permanently store data 

#### vocabulary 

- RDBMS: The software that manages the data, i.e., handles it's storage and retrieval. Runs the database server

- Database: Sometimes used interchangably with a DBMS; The actual location of the data stored on disk; You won't interact with this directly, but rather, through the DBMS.

- Database Client: a program that can be used to connect to a database. Each database usually comes with an officially sanctioned command line client, and there are many different GUI database clients as well.
    - MySQL Workbench is a database client and design tool
    
- Database Server: a computer that runs the DBMS and stores the data, typically either on-premises or in the cloud.

- DDL: Data definition language; commands that change the structure of the database or the DBMS itself, or change the structure of tables in the database. We will not be focusing on this part of SQL.

- DML: Data manipulation language; used to insert, update, delete, and retrieve information from databases. We will be focusing on the the retrieval part of this.


#### queries 

- we use structured query langauge (SQL) to interact with MySQL
- sql is madeup of statements and commands sent to the server individually with results sent back to the database client (MySQL Workbench)

## MySQL Clauses

"clauses" are stipulations that are added to queries to determine which rows are targeted 


"WHERE" clauses

*example* syntax

- select * FROM <table_name> where <some_condition> is true;
    - example: select * FROM employees where birth_year >= "1970";

*pattern matching*

enables you to use _ to match any single character and % to match an arbitrary number of characters (including zero characters). In MySQL, SQL patterns are case-insensitive by default

"%" is considered a 'wildcard' 

*example* syntax

select DISTINCT <column> FROM <table> WHERE <column> LIKE '%<condition%'>;

    - example: SELECT DISTINCT <first_name> FROM <employees> WHERE <first_name> LIKE '%sus%';


*keep in mind that in SQL there is no pure 'boolean' value; instead it recognizes "FALSE" as 0 and "TRUE" as 1*


exmaple of pattern matching: To find all the employees with the last name of 'Herber', 'Dredge', 'Lipner', and 'Baek':


SELECT <emp_no>, <first_name>, <last_name>
FROM <employees>
WHERE <last_name> IN ('Herber', 'Dredge', 'Lipner', 'Baek');


### Chaining WHERE clauses

- we can chain together an IN clause with a LIKE clause, or any of the clauses, using AND and OR

    - for example:
            SELECT emp_no, first_name, last_name
            FROM employees
            WHERE last_name IN ('Herber','Baek')
            AND emp_no < 20000;

## Date and Time functions

### MONTH function

- eg. select month('2022-05-19');

- *rule of thumb* SQL mondays start at num '0'

### DAY function

- eg. select *, dayname(hire_date) as day_hired from employees;

### CURDATE function 

- function returns just the current date with time information in "YYYY-MM-DD" format 

- however, keep in mind that this will be configured to the administrative sql database server (a.k.a. where the host server is located)

### DATEDIFF function 

- returns the difference of two defined dates in number of hours, days, weeks, years, etc.)

- eg. select round(datediff(curdate(), hire_date)/365 as tenure from employees;

### UNIX_TIMESTAMP & UNIX_TIMESTAMP(date)

- is used to represent time as an integer. It will return the number of seconds since midnight January 1st, 1970

- if you pass a date time value to UNIX_TIMESTAMP(), it will give you the number of seconds from the unix epoch to that date
